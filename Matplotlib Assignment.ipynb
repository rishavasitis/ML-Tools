Q1: **Matplotlib and Its Usage:**
   - Matplotlib is a Python library for creating static, animated, or interactive visualizations in a variety of formats. It's used to display data in a graphical form. Five plots that can be plotted using the Pyplot module of Matplotlib include line plots, bar plots, scatter plots, histograms, and pie charts.

Q2: **Creating a Scatter Plot:**
   ```python
   import matplotlib.pyplot as plt
   import numpy as np

   np.random.seed(3)
   x = 3 + np.random.normal(0, 2, 50)
   y = 3 + np.random.normal(0, 2, len(x))

   plt.scatter(x, y)
   plt.title("Scatter Plot")
   plt.xlabel("X-axis")
   plt.ylabel("Y-axis")
   plt.show()
   ```

Q3: **Usage of subplot() Function:**
   - The `subplot()` function is used to create multiple plots in a single figure. It divides the figure into rows and columns of subplots. You can draw four line plots as specified using the `subplot()` function.
   ```python
   import matplotlib.pyplot as plt
   import numpy as np

   x = np.array([0, 1, 2, 3, 4, 5])
   y1 = np.array([0, 100, 200, 300, 400, 500])
   y2 = np.array([50, 20, 40, 20, 60, 70])
   y3 = np.array([10, 20, 30, 40, 50, 60])
   y4 = np.array([200, 350, 250, 550, 450, 150])

   plt.subplot(2, 2, 1)
   plt.plot(x, y1)
   plt.title("Line Plot 1")

   plt.subplot(2, 2, 2)
   plt.plot(x, y2)
   plt.title("Line Plot 2")

   plt.subplot(2, 2, 3)
   plt.plot(x, y3)
   plt.title("Line Plot 3")

   plt.subplot(2, 2, 4)
   plt.plot(x, y4)
   plt.title("Line Plot 4")

   plt.tight_layout()
   plt.show()
   ```

Q4: **Bar Plot and Its Usage:**
   - A bar plot is used to display data with rectangular bars. It's often used to compare data between different categories. Here's how to plot a bar plot and a horizontal bar plot:
   ```python
   import matplotlib.pyplot as plt
   import numpy as np

   company = np.array(["Apple", "Microsoft", "Google", "AMD"])
   profit = np.array([3000, 8000, 1000, 10000])

   # Vertical Bar Plot
   plt.bar(company, profit)
   plt.title("Vertical Bar Plot")
   plt.xlabel("Company")
   plt.ylabel("Profit")
   plt.show()

   # Horizontal Bar Plot
   plt.barh(company, profit)
   plt.title("Horizontal Bar Plot")
   plt.xlabel("Profit")
   plt.ylabel("Company")
   plt.show()
   ```

Q5: **Box Plot and Its Usage:**
   - A box plot is used to visualize the distribution of a dataset and identify potential outliers. It shows the median, quartiles, and potential outliers. To plot a box plot:
   ```python
   import matplotlib.pyplot as plt
   import numpy as np

   box1 = np.random.normal(100, 10, 200)
   box2 = np.random.normal(90, 20, 200)

   data = [box1, box2]
   plt.boxplot(data)
   plt.title("Box Plot")
   plt.xlabel("Data Sets")
   plt.ylabel("Values")
   plt.show()
   ```
