Q1. **Difference in Data Type of list_ and array_list:**
   - Yes, there is a difference in data type. `list_` contains strings, while `array_list` contains integers after conversion. To print their data types, use the following code:
   ```python
   print(type(list_))
   print(type(array_list))
   ```

Q2. **Data Type of Each Element in list_ and array_list:**
   - To print the data type of each element in both variables, you can loop through the elements and use the `type()` function. Here's an example:
   ```python
   for item in list_:
       print(type(item))
   for element in array_list:
       print(type(element))
   ```

Q3. **Difference in Data Types with dtype=int:**
   - When you change the data type of `array_list` to integer using `dtype=int`, it will contain integers. To print data types of elements:
   ```python
   for item in list_:
       print(type(item))
   for element in array_list:
       print(type(element))
   ```
