Q2. **Difference between loc and iloc in pandas.DataFrame:**
   - `loc` is like using labels (column names or row indexes) to access data, while `iloc` is like using integer positions. For example, `df.loc[1]` would access the row labeled '1', whereas `df.iloc[1]` would access the second row (position 1) in the DataFrame.

   ```python
   # Now, accessing rows in new_df:
   print(new_df.loc[2])   # Accessing the row with label '2'.
   print(new_df.iloc[2])  # Accessing the third row (position 2).
   ```

   - This code calculates the difference between two dates in days, hours, and minutes using Pandas time delta:
   ```python

   date1 = pd.to_datetime(input('Enter the first date (YYYY-MM-DD): '))
   date2 = pd.to_datetime(input('Enter the second date (YYYY-MM-DD): '))
   time_diff = date2 - date1
   print(f'Difference in Days: {time_diff.days}')
   ```


Q11. **Calculate Mean, Median, and Mode of Test Scores:**
   - This code calculates the mean, median, and mode of test scores in a CSV file and displays the results in a table:
   ```python
   import pandas as pd
   file_path = input('Enter the file path: ')
   df = pd.read_csv(file_path)
   mean = df['Test Score'].mean()
   median = df['Test Score'].median()
   mode = df['Test Score'].mode()
   results = pd.DataFrame({'Statistic': ['Mean', 'Median', 'Mode'], 'Value': [mean, median, ', '.join(map(str, mode))]})
   print(results)
   ```
