Q1. **Five Functions of Pandas:**
   - 1. `pd.read_csv()`: Used to read data from a CSV file.
   - 2. `df.head()`: Displays the first few rows of a DataFrame.
   - 3. `df.describe()`: Provides summary statistics of the DataFrame.
   - 4. `df.drop()`: Removes rows or columns from a DataFrame.
   - 5. `df.groupby()`: Groups data for analysis.

Q2. **Re-indexing a Pandas DataFrame:**
   ```python
   def reindex_dataframe(df):
       df.index = range(1, 2*len(df) + 1, 2)
       return df
   ```

Q3. **Calculating Sum of First Three Values in 'Values' Column:**
   ```python
   def sum_first_three_values(df):
       total = df['Values'].head(3).sum()
       print(total)
   ```

Q4. **Creating 'Word_Count' Column:**
   ```python
   def count_words(df):
       df['Word_Count'] = df['Text'].apply(lambda x: len(x.split()))
       return df
   ```

Q5. **Difference Between DataFrame.size() and DataFrame.shape():**
   - `DataFrame.size` returns the total number of elements in the DataFrame.
   - `DataFrame.shape` returns a tuple representing the dimensions (number of rows and columns).

Q6. **Function to Read an Excel File in Pandas:**
   - You use `pd.read_excel('file.xlsx')` to read an Excel file.

Q7. **Extracting Username from Email Addresses:**
   ```python
   def extract_username(df):
       df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
       return df
   ```

Q8. **Selecting Rows Based on Conditions:**
   ```python
   def select_rows(df):
       selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
       return selected_rows
   ```

Q9. **Calculating Mean, Median, and Standard Deviation:**
   ```python
   def calculate_stats(df):
       mean = df['Values'].mean()
       median = df['Values'].median()
       std_dev = df['Values'].std()
       return mean, median, std_dev
   ```

Q10. **Calculating Moving Average:**
   ```python
   def calculate_moving_average(df):
       df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
       return df
   ```

Q11. **Adding 'Weekday' Column:**
   ```python
   def add_weekday(df):
       df['Weekday'] = pd.to_datetime(df['Date']).dt.day_name()
       return df
   ```

Q12. **Selecting Rows Between Specific Dates:**
   ```python
   def select_dates(df):
       start_date = '2023-01-01'
       end_date = '2023-01-31'
       selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
       return selected_rows
   ```
