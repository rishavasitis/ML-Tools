Q1: **Plots Using Seaborn and Their Uses:**
   - 1. **Scatter Plot:** Used to display the relationship between two continuous variables.
   - 2. **Bar Plot:** Shows comparisons among categories or discrete data.
   - 3. **Box Plot:** Displays the distribution, median, quartiles, and potential outliers.
   - 4. **Violin Plot:** Combines a box plot with a kernel density estimation to show the data distribution.
   - 5. **Pair Plot:** Useful for exploring relationships between multiple variables in a dataset.

Q2: **Line Plot with "fmri" Dataset:**
   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   fmri = sns.load_dataset("fmri")
   sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
   plt.show()
   ```

Q3: **Box Plots with "titanic" Dataset:**
   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   titanic = sns.load_dataset("titanic")
   sns.boxplot(x="pclass", y="age", data=titanic)
   plt.show()

   sns.boxplot(x="pclass", y="fare", data=titanic)
   plt.show()
   ```

Q4: **Histogram with "diamonds" Dataset:**
   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   diamonds = sns.load_dataset("diamonds")
   sns.histplot(data=diamonds, x="price", hue="cut", element="step", common_norm=False)
   plt.show()
   ```

Q5: **Pair Plot with "iris" Dataset:**
   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   iris = sns.load_dataset("iris")
   sns.pairplot(data=iris, hue="species")
   plt.show()
   ```

Q6: **Heatmap with "flights" Dataset:**
   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   flights = sns.load_dataset("flights")
   pivot_table = flights.pivot_table(index="month", columns="year", values="passengers")
   sns.heatmap(pivot_table, cmap="YlGnBu")
   plt.show()
   ```
